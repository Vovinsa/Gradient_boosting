```python
from sklearn.datasets import make_regression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
import pandas as pd
```

```python
X, y = make_regression(n_samples=20, n_features=5)
```

```python
df = pd.DataFrame(X)
df["y_true"] = y

df
```

```python
lr = 0.1
n_trees = 1000
df["y_pred"] = df["y_true"].mean()
trees = []

for _ in range(n_trees):
    df["residual"] = df["y_true"] - df["y_pred"]
    tree = DecisionTreeRegressor(max_depth=1)
    tree.fit(df.drop(["y_true", "y_pred", "residual"], axis=1), df["residual"])
    df["y_pred"] += lr * tree.predict(df.drop(["y_true", "y_pred", "residual"], axis=1))
    trees.append(tree)
    print(mean_squared_error(df["y_true"], df["y_pred"]))
```

```python
test = df[[0, 1, 2, 3, 4]].copy()
test
```

```python
test["y_pred"] = df["y_true"].mean()

for tree in trees:
    test["y_pred"] += lr * tree.predict(test.drop("y_pred", axis=1))
```
